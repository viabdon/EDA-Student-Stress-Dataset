```python
# Distribuição cruzada
sns.boxplot(data=stress_lvl, x="sleep_quality", y="academic_performance", hue="breathing_problem")
plt.title("Sono e problemas respiratórios vs performance acadêmica")
plt.show()

sns.boxplot(data=stress_lvl, x="sleep_quality", y="stress_level", hue="breathing_problem")
plt.title("Sono e problemas respiratórios vs nível de estresse")
plt.show()

# Correlações
stress_lvl[['sleep_quality', 'breathing_problem', 'academic_performance']].corr()
stress_lvl[['sleep_quality', 'breathing_problem', 'stress_level']].corr()
```

```python
# Comparação de médias
sns.barplot(data=stress_lvl, x="social_support", y="stress_level", hue="extracurricular_activities")
plt.title("Suporte social e lazer vs nível de estresse")
plt.show()

sns.barplot(data=stress_lvl, x="social_support", y="anxiety_level", hue="extracurricular_activities")
plt.title("Suporte social e lazer vs nível de ansiedade")
plt.show()

# Agrupamento
stress_lvl.groupby(["social_support", "extracurricular_activities"])[["stress_level", "anxiety_level"]].mean()
```

```python
# Dispersão
sns.scatterplot(data=stress_lvl, x="study_load", y="anxiety_level", hue="future_career_concerns")
plt.title("Carga de estudo e preocupações futuras vs ansiedade")
plt.show()

# Regressão
sns.lmplot(data=stress_lvl, x="study_load", y="anxiety_level", hue="future_career_concerns")
plt.title("Relação study_load e ansiedade com preocupações futuras")
plt.show()

# Agrupamento
stress_lvl.groupby("future_career_concerns")["anxiety_level"].mean()
```
